```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() in another format
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes() # stopping here returns a blank chart
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show() #looks for all active figure objects to display
            #...can cause problems so once per session
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with color and legend for Boston.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Fare($)", fontweight="bold")
plt.ylabel("Date", fontweight="bold")
# Invert y-axis so January is at the top.
plt.gca().invert_yaxis()
# Create a title.
plt.title("PyBer Fare by Month", fontweight="bold")
# Add the legend.
legend_properties = {'weight':'bold'}
plt.legend(prop=legend_properties)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Scatter plot - code option 1
plt.plot(x_axis, y_axis, 'o')
```

```python
# Scatter plot - code option 2
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
# Create title and axis labels.
plt.title("PyBer Fare by Month", fontweight="bold")
plt.xlabel("Fare($)", fontweight="bold")
plt.ylabel("Date", fontweight="bold")
# Invert y-axis data
plt.gca().invert_yaxis()
# Set limit for x-axis data between 0 and 51.
plt.xlim(0, 51)
# Add legend.
legend_properties = {'weight':'bold'}
plt.legend(prop=legend_properties)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Change/ adjust marker size.
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# List comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Scatter plot with object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Bubble chart with object-oriented interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], 
           c="deepskyblue", edgecolor="black", alpha=.8, 
           lw=2, label="Boston")
# Create title and axis labels.
plt.title("PyBer Fare by Month", fontweight="bold")
plt.xlabel("Fare($)", fontweight="bold")
plt.ylabel("Date", fontweight="bold")
# Add a limit for x-axis data (between 0 and 51).
plt.xlim(0, 51)
# Invert y-axis data (January on top)
plt.gca().invert_yaxis()
# Add legend.
legend_properties = {'weight':'bold'}
plt.legend(prop=legend_properties)
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    shadow=3,
    autopct='%.1f%%')

plt.show()
```
